```python
import pandas as pd
df=pd.read_csv('/content/iris-write-from-docker (1).csv')
print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
```

```python
import matplotlib.pyplot as plt
x=df['petal_length']
y=df['petal_width']
plt.bar(x,y)
plt.title('bar graph')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=df['petal_length']
y=df['petal_width']
plt.hist(x)
plt.title('histogram graph')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=df['petal_length']
y=df['petal_width']
plt.scatter(x,y)
plt.title('scatter graph')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=df['petal_length']
y=df['petal_width']
plt.boxplot(x)
plt.title('boxplot graph')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.show()
```
