```python
import matplotlib.pyplot as plt
# Sample data
categories = ['Category A', 'Category B', 'Category C']
amounts = [25, 50, 30]
# Create a bar chart
plt.bar(categories, amounts)
# Adding labels
plt.xlabel('Categories')
plt.ylabel('Amounts')
plt.title('Bar Chart of Amounts by Category')
# Show the plot
plt.show()
```
